```python
import torch
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

```python
model=torch.hub.load('ultralytics/yolov5','yolov5s')
```

```python
model
```

```python
img='https://ultralytics.com/images/zidane.jpg'
```

```python
results =model(img)
results.print()
```

```python
%matplotlib inline
plt.imshow(np.squeeze(results.render()))
plt.show()
```

```python
cap = cv2.VideoCapture(0)
while cap.isOpened():
    ret, frame = cap.read()
    
    # Make detections 
    results = model(frame)
    
    cv2.imshow('Drowsiness Detector', np.squeeze(results.render()))
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
